```python
import pandas as pd
origin = pd.read_csv('Comparison.csv')
compare = origin.copy()
```

```python
origin
```

```python
compare['MACD_max'] = compare.groupby(['Time Period (% ago)'])['MACD ROI'].transform(max)
compare['Heikin-ashi_max'] = compare.groupby(['Time Period (% ago)'])['Heikin-ashi ROI'].transform(max)
compare['pro_heikin-ashi_max'] = compare.groupby(['Time Period (% ago)'])['Processed-Heikin-ashi ROI'].transform(max)
compare['heikin_MACD_max'] = compare.groupby(['Time Period (% ago)'])['Heikin-ashi plus MACD ROI'].transform(max)
```

```python
compare
```

```python
mean_value = compare.groupby(['Time Interval']).mean()[['MACD ROI','Heikin-ashi ROI','Processed-Heikin-ashi ROI','Heikin-ashi plus MACD ROI']]
```
