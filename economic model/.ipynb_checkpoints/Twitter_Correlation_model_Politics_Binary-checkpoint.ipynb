```python
import pymongo
from sklearn.externals import joblib
import numpy as np
import pickle
import pprint
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
def get_probability(data):
    str_test = data
    x_test = [str_test]
    x_test = np.asarray(x_test)
    hash_vec = pickle.load(open("model/subjectClassifierNBHashAutoBigram.pkl", 'rb'))

    tf = pickle.load(open("model/subjectClassifierNBTfidfAutoBigram.pkl", 'rb'))
    x_test_hash = hash_vec.transform(x_test)
    x_test_tfidf = tf.transform(x_test_hash)
    clf = joblib.load('model/subjectClassifierNBAutoBigram.pkl')
    prediction_score = clf.predict_proba(x_test_tfidf).tolist()
    predicted_class = clf.classes_.tolist()
    prediction = dict(zip(predicted_class, prediction_score[0]))
    highest = sorted(prediction, key=prediction.get, reverse=True)[0]
    return prediction
```

```python
client = pymongo.MongoClient(host='localhost', port=27017)
db = client.frank
tweets = db.tweets
```

```python
def calculate_correlation_score(username):
    correlation_matrix = pd.DataFrame()
    SentenceID = []
    Joy = []
    Anger = []
    Sad = []
    Fear = []
    Disgust = []
    Text = []
    num = 1
    for i in tweets.find({'UserName':username}):
        SentenceID.append(num)
        Joy.append(i['Joy'])
        Anger.append(i['Anger'])
        Sad.append(i['Sad'])
        Fear.append(i['Fear'])
        Disgust.append(i['Disgust'])
        Text.append(i['Text'])
        num+=1
    
    correlation_matrix['SentenceID'] = SentenceID
    correlation_matrix['Text'] = Text
    correlation_matrix['Joy'] = Joy
    correlation_matrix['Anger'] = Anger
    correlation_matrix['Sad'] = Sad
    correlation_matrix['Fear'] = Fear
    correlation_matrix['Disgust'] = Disgust
    
    
    entertainment = []
    finance = []
    food = []
    health = []
    others = []
    politics = []
    scitech = []
    sports = []
    travel = []
    videogames = []

    for i in range(len(correlation_matrix)):
        text = correlation_matrix['Text'][i]
        prob_dict = get_probability(text)
        entertainment.append(prob_dict['entertainment'])
        finance.append(prob_dict['finance'])
        food.append(prob_dict['food'])
        health.append(prob_dict['health'])
        others.append(prob_dict['others'])
        politics.append(prob_dict['politics'])
        scitech.append(prob_dict['scitech'])
        sports.append(prob_dict['sports'])
        travel.append(prob_dict['travel'])
        videogames.append(prob_dict['videogames'])

    correlation_matrix['Entertainment'] = entertainment
    correlation_matrix['Finance'] = finance
    correlation_matrix['Food'] = food
    correlation_matrix['Health'] = health
    correlation_matrix['Others'] = others
    correlation_matrix['Politics'] = politics
    correlation_matrix['Scitech'] = scitech
    correlation_matrix['Sports'] = sports
    correlation_matrix['Travel'] = travel
    correlation_matrix['Videogames'] = videogames    
    
    Joy = correlation_matrix['Joy'].tolist()
    Anger = correlation_matrix['Anger'].tolist()
    Sad = correlation_matrix['Sad'].tolist()
    Fear = correlation_matrix['Fear'].tolist()
    Disgust = correlation_matrix['Disgust'].tolist()
    Entertainment = correlation_matrix['Entertainment'].tolist()
    Finance = correlation_matrix['Finance'].tolist()
    Food = correlation_matrix['Food'].tolist()
    Health = correlation_matrix['Health'].tolist()
    Others = correlation_matrix['Others'].tolist()
    Politics = correlation_matrix['Politics'].tolist()
    Scitech = correlation_matrix['Scitech'].tolist()
    Sports = correlation_matrix['Sports'].tolist()
    Travel = correlation_matrix['Travel'].tolist()
    Videogames = correlation_matrix['Videogames'].tolist()
    
    return correlation_matrix
```

```python
correlation = calculate_correlation_score('donaldtrump')
```

```python
data = []
for i in politics_column:
    data.append(correlation.iloc[i].tolist())
```

```python
filter_politics = pd.DataFrame(data,columns=['Index','Text','Joy','Anger','Sad','Fear','Disgust','Entertainment','Finance','Food','Health','Others','Politics','Scitech','Sports','Travel','Videogames'])
```

```python
emotions=filter_politics[['Joy','Anger','Sad','Fear','Disgust']]
max_emotions=emotions.max(axis=1)
for i in range(len(max_emotions)):
    for j in range(len(emotions.iloc[i])):
        if max_emotions[i]==emotions.iloc[i,j]:
            emotions.iloc[i,j]=1
        else:
            emotions.iloc[i,j]=0
emotions
a = filter_politics[['Text','Entertainment','Finance','Food','Health','Others','Politics','Scitech','Sports','Travel','Videogames']]
new_filter_politics=pd.concat([a,emotions],axis=1, sort=False)
```

```python
new_filter_politics
```

```python
joy_politics_corr = []
anger_politics_corr = []
sad_politics_corr = []
fear_politics_corr = []
disgust_politics_corr = []
```

```python
filter_politics = new_filter_politics.copy()
length = len(filter_politics)
for i in range(2,length):
    joy_politics_corr.append(pearsonr(filter_politics['Joy'][:i], filter_politics['Politics'][:i])[0])
    anger_politics_corr.append(pearsonr(filter_politics['Anger'][:i], filter_politics['Politics'][:i])[0])
    sad_politics_corr.append(pearsonr(filter_politics['Sad'][:i], filter_politics['Politics'][:i])[0])
    fear_politics_corr.append(pearsonr(filter_politics['Fear'][:i], filter_politics['Politics'][:i])[0])
    disgust_politics_corr.append(pearsonr(filter_politics['Disgust'][:i], filter_politics['Politics'][:i])[0])
```

```python
correlation = pd.DataFrame()
index = [i for i in range(1,length-1)]
correlation['index'] = index
correlation['joy/politics'] = joy_politics_corr
correlation['anger/politics'] = anger_politics_corr
correlation['sad/politics'] = sad_politics_corr
correlation['fear/politics'] = fear_politics_corr
correlation['disgust/politics'] = disgust_politics_corr
correlation
```

```python
fig = plt.figure(figsize=(20,10))
x = correlation['index'].values.tolist()
joy_politics = correlation['joy/politics'].values.tolist()
anger_politics = correlation['anger/politics'].values.tolist()
sad_politics = correlation['sad/politics'].values.tolist()
fear_politics = correlation['fear/politics'].values.tolist()
disgust_politics = correlation['disgust/politics'].values.tolist()
plt.plot(x,joy_politics,color = 'red',label = 'joy_politics')
plt.plot(x,anger_politics,color = 'blue',label = 'anger_politics')
plt.plot(x,sad_politics,color = 'green',label = 'sad_politics')
plt.plot(x,fear_politics,color = 'yellow',label = 'fear_politics')
plt.plot(x,disgust_politics,color = 'gray',label = 'disgust_politics')
plt.xlabel('sentence quantity',fontsize = 20)
plt.ylabel('correlation score',fontsize = 20)
plt.legend()
```
