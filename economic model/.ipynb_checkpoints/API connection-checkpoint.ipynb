```python
import json
url_moji = "https://0ujrqdyqc6.execute-api.ap-southeast-2.amazonaws.com/test/mojiresource"
param_moji = '{"id": "1", "text" : "The faces a local domestic legislator, and this is not an election campaign gambit. It is a very public gesture of disdain.", "userId": "123" , "timestamp" : "2019/10/30", "applications" : "fb"}' 
moji = requests.post(url_moji, param_moji) 
data = moji.json()['stdout']
parsedData = json.loads(data)
parsedData
```

```python
import pandas as pd
label_pair = []
subject_dataframe = pd.DataFrame()
label_list = []
intensity_list = []
length = int(len(parsedData))
for i in range(length-5,length):
    label = label_list.append(parsedData[i]['Emotion_label'])
    intensity = intensity_list.append(parsedData[i]['Intensity'])
subject_dataframe['label'] = label_list
subject_dataframe['intensity'] = intensity_list
subject_dataframe
```

```python
score = 0
for i in range(len(subject_dataframe)):
    score += subject_dataframe['intensity'][i]
score
```

```python
label = "Anger"
highest_score = subject_dataframe['intensity'][0]
for i in range(len(subject_dataframe)):
    if(subject_dataframe['intensity'][i] > highest_score):
        label = subject_dataframe['label'][i]
        highest_score = subject_dataframe['intensity'][i]
label_pair.append(label)
label_pair.append(highest_score)
label_pair
```
